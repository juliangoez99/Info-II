```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0, 3 * np.pi, 200)# definir el tipo de grafica
y1 = np.sin (x1)
x2 = np.linspace(0, 2 * np.pi, 200)#define el tipo de grafica
y2 = np.cos (x2)
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(6,9))#fenife cantidad de graficas y el tamaño de estas
ax1.plot(x1, y1, label = "grafica seno",color="red",linewidth=1,linestyle="--") #muesta lo que contiene la grafica como ejes, y nombre de graficacolor= "cambia el color de linea , linewidth= el ancho de la linea,  linestyle= "--",":"
ax2.plot(x2, y2, label = "grafica cos", color="green", linewidth=4)
ax1.set_xlabel("eje x")#nombra eje x
ax1.set_ylabel("eje y")#nombra eje y
ax1.set_title("grafiaca de la funcion Seno(x)")#tittulo de la grafica
ax1.legend()
ax1.grid() #agrega cuadricula
ax1.annotate("local max", xy =(1.5 ,1), xytext=(3,1.5), arrowprops=dict(facecolor="red", shrink=0.05)) #agregar una flecha
ax1.text(3.5,0.5,"prueba") #genera texxto en la grafica
ax2.set_xlabel("eje x")
ax2.set_ylabel("eje y")
ax2.set_title("grafiaca de la funcion Seno(x)")
ax2.legend()
ax2.grid()
plt.show#mostar la grafica


np.max(y1)
np.max(y2)
```
